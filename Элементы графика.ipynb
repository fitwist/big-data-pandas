```python
import numpy as np
from numpy.random import randn

import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
# %matplotlib inline указывает, что график необходимо построить все в той же оболочке Jupyter,
# но теперь он выводится как обычная картинка.
%matplotlib inline
rcParams['figure.figsize'] = 5,4 # размер холста
```

```python
# Зададим координаты
x = range(1, 10) 
y = [1, 2, 3, 4, 5, 0, 4, 3, 2]

fig = plt.figure() # Создание пустой фигуры
ax = fig.add_axes([.1, .1, 1, 1]) # Добавление осей, x-y-x-y, немного смещаем их
ax.plot(x,y)
```

```python
# Объектно-ориентированный метод построения графиков
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

# Передадим массив в качестве границ отображения оси x, y
ax.set_xlim([1,9]) 
ax.set_ylim([0,5])

# Мы хотим только определенные значения на осях, зададим их
ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,4,5])

ax.plot(x, y)
```

```python
# Переиспользуем код
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.set_xlim([1,9]) 
ax.set_ylim([0,5])

# Простая сетка для улучшения читаемости и категоризации
ax.grid()
ax.plot(x, y)
```

```python
# Сгенерируем субграфики
fig = plt.figure() # Создадим пустой объект
fig, (ax1, ax2) = plt.subplots(1, 2) # В первом графике 1 строка, 2 столбца

# Преобразуем один график в два, причем первый график, получив только координату x, экстраполирует значения y, 
# присвоив им порядковые номера элементов массива x
ax1.plot(x) 
ax2.plot(x, y)
```
